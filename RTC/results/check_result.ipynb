```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('RTC/results/sim_result.csv', index_col=0, parse_dates=True, date_format='%d/%m %H:%M')

df = df.T

# Number of bars per group
n_bars = len(df.columns)
# Number of groups (original columns)
n_groups = len(df.index)

# Create a figure and an axis
fig, ax = plt.subplots(figsize=(15,7))

# Bar width
bar_width = 0.03

# Positions of the groups on the x-axis
indices = np.arange(n_groups)
# Plotting the bars
for i, column in enumerate(df.columns):
    bar_positions = indices + i * bar_width
    ax.bar(bar_positions, df[column], bar_width, label=column)

# Set the position of the x ticks
ax.set_xticks(indices + bar_width * (n_bars - 1) / 2)
ax.set_xticklabels(df.index)

# Adding labels and title
ax.set_xlabel('Obj. Func. Contributor')
ax.set_ylabel('Objective Function value')
ax.set_title('Comparison per itteration')
# ax.legend(title='Index')

# Show the plot
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('RTC/results/full_sim_result.csv', index_col=0, parse_dates=True, date_format='%d/%m %H:%M')

df = df.T

# Number of bars per group
n_bars = len(df.columns)
# Number of groups (original columns)
n_groups = len(df.index)

# Create a figure and an axis
fig, ax = plt.subplots(figsize=(15,7))

# Bar width
bar_width = 0.03

# Positions of the groups on the x-axis
indices = np.arange(n_groups)
# Plotting the bars
for i, column in enumerate(df.columns):
    bar_positions = indices + i * bar_width
    ax.bar(bar_positions, df[column]  * [1,1,1,1,1,500,500,1,1], bar_width, label=column)

# Set the position of the x ticks
ax.set_xticks(indices + bar_width * (n_bars - 1) / 2)
ax.set_xticklabels(df.index)

# Adding labels and title
ax.set_xlabel('Obj. Func. Contributor')
ax.set_ylabel('Objective Function value')
ax.set_title('Comparison per itteration of spill volume roughly')
# ax.legend(title='Index')

# Show the plot
plt.show()
```
